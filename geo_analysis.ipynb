```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
outline_border = gpd.read_file(
    "./geodata/small_bangladesh_geojson_adm0_whole_bangladesh_outline.json"
)
divisions = gpd.read_file(
    "./geodata/small_bangladesh_geojson_adm1_8_divisions_bibhags.json"
)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax = outline_border.plot(ax=ax, edgecolor="red", linewidth=2)
ax.set_title("Border of Bangladesh")
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax = divisions.plot(ax=ax, edgecolor="red", linewidth=2)
ax.set_title("Division of Bangladesh with area specification")
```

```python
# Divisions name
division_name = divisions.ADM1_EN
division_name
```

```python
division_infected = [
    1,
    2,
    5 + 17,
    12 + 13 + 1 + 1 + 1,
    9 + 2,
    1 + 10 + 29 + 14 + 3 + 12,
    1 + 12 + 6 + 9 + 24 + 3 + 16,
    1,
]
```

```python
divisions["Infected"] = division_infected
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax = divisions.plot(
    ax=ax, edgecolor="black", column="Infected", markersize=10, legend=True
)

divisions.apply(
    lambda x: ax.annotate(
        text=x.ADM1_EN, xy=x.geometry.centroid.coords[0], ha="center"
    ),
    axis=1,
)
ax.set_title("Name of Divisions in Bangladesh")
```

```python
districts = ['Panchagarh', 'Thakurgaon', 'Nilphamari', 'Dinajpur', 'Lalmonirhat', 'Kurigram', 'Rangpur', 'Gaibandha', 'Joypurhat',
                 'Naogaon', 'Rajshahi', 'Bogra', 'Natore', 'Mymensingh', 'Tangail', 'Meherpur', 'Kushtia', 
                 'Pabna', 'Manikganj', 'Dhaka', 'Chuadanga', 'Jhenaidah', 'Magura', 'Faridpur', 'Narsingdi', 'Comilla', 'Narail',
                 'Gopalganj', 'Madaripur', 'Shariatpur', 'Khulna', 'Jhalokati']
```

```python
infected = [1, 12, 6, 9, 24, 3, 16, 1, 10, 29, 14, 3, 9, 2, 12, 13, 12, 5, 17, 1, 1, 2, 6, 71, 1, 1, 1, 8, 6, 3, 1, 1]
```

```python
sum(infected)
```

```python
d = {k: v for (k, v) in zip(districts, infected)}
```

```python
d
```

```python
districts_zillas = gpd.read_file(
    "./geodata/small_bangladesh_geojson_adm2_64_districts_zillas.json"
)
```

```python
districts_zillas
```

```python
district_names = districts_zillas.ADM2_EN
```

```python
district_names
```

```python
infected = []
for name in district_names:
    if name in districts:
        infected.append(d[name])
    else:
        infected.append(0)

infected
```

```python
districts_zillas["Infected"] = infected
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax = districts_zillas.plot(
    ax=ax, edgecolor="black", column="Infected", markersize=10, legend=True
)

districts_zillas.apply(
    lambda x: ax.annotate(
        text=x.ADM2_EN, xy=x.geometry.centroid.coords[0], ha="center"
    ),
    axis=1,
)
ax.set_title("Heatmap of Infected by Zilla")
```
